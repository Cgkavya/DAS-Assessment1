```python
# Check for and remove any duplicated rows.

df = df.drop_duplicates()
```

```python
# Handle missing values if any: Fill missing numerical features with the mean value.

df['MedInc'] = df['MedInc'].fillna(df['MedInc'].mean())
df['HouseAge'] = df['HouseAge'].fillna(df['HouseAge'].mean())
df['AveRooms'] = df['AveRooms'].fillna(df['AveRooms'].mean())
df['AveBedrms'] = df['AveBedrms'].fillna(df['AveBedrms'].mean())
df['Population'] = df['Population'].fillna(df['Population'].mean())
df['AveOccup'] = df['AveOccup'].fillna(df['AveOccup'].mean())
df['Latitude'] = df['Latitude'].fillna(df['Latitude'].mean())
df['Longitude'] = df['Longitude'].fillna(df['Longitude'].mean())
df['MedHouseVal'] = df['MedHouseVal'].fillna(df['MedHouseVal'].mean())
```

```python
# Create a new column PricePerRoom = median_house_value / total_rooms.

df['PricePerRoom'] = df['MedHouseVal'] / df['AveRooms']
```

```python
#  Create a column HighPopulationArea: 1 if population > 500, else 0

df['HighPopulationArea'] = [1 if x > 500 else 0 for x in df['Population']]
```

```python
# Bin the median_income into 5 equal-sized bins and label them as Very Low, Low, Medium,High, Very High.

bins = [0, 20, 40, 60, 80, 100]
lab = ['Very Low', 'Low', 'Medium', 'High', 'Very High']
df["IncomeCategories"] = pd.cut(df['MedInc'], bins = bins, labels = lab, include_lowest = True)
```

```python
# Plot the distribution of median_house_value with a histogram.

plt.hist(df['MedHouseVal'], color = "green", rwidth = 0.8, density = True)
plt.xlabel('X-axis')
plt.ylabel('Y-Axis')
plt.title('Histogram of median_house_value')
plt.show()
```

```python
# Create a scatter plot of longitude vs latitude, colored by median_house_value.

plt.figure(figsize=(5,3.5))
plt.scatter(df['Latitude'], df['Longitude'], c = df['MedHouseVal'], marker = "*")
plt.xlabel('Latitude')
plt.ylabel('Longitude')
plt.title('Latitude vs Longitude')
plt.show()
```

```python
#  Plot a boxplot of median_house_value grouped by the new income categories

import seaborn as sns 
sns.boxplot (x = df['MedHouseVal'], 
            hue = df['IncomeCategories'])
```

```python
#  Plot the correlation matrix heatmap between numerical features

co_mtx = df.corr(numeric_only=True)
print(co_mtx)
sns.heatmap(co_mtx, cmap="YlGnBu", annot=True)
plt.show()
```
